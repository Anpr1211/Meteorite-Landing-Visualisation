```python
from geojson import Point, Feature, FeatureCollection, dump
```

```python
features=[]
```

```python
for i in d.keys():
    for j in range(len(d[i])):
        n=list(d[i][j].keys())[0]
        area_n=i
        lat=d[i][j][n]["Lat"]
        lng=d[i][j][n]["Lng"]   
        point=Point((lng,lat))
        features.append(Feature(geometry=point,properties={"area_name":area_n,"name":n}))
```

```python
len(features)
```

```python
feature_collection = FeatureCollection(features)
```
