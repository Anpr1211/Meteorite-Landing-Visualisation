```python
import numpy
import pandas
from scipy import stats
import pert
import matplotlib.pyplot as plt
```

```python
num_sim = 10000

x = (5, 7, 8, 9)
p = (0.3, 0.5, 0.1, 0.1)

npv_ = []
```

```python
for i in range(0, num_sim):
    
    price = round(numpy.random.triangular(15, 19, 23), 2)

    material = round(pert.PERT(1, 2, 3).rvs()[0], 2)

    labour = stats.rv_discrete(values=(x, p)).rvs()

    overhead = round(numpy.random.normal(0.5, 0.01), 2)
    
    # cashflow for year 1
    cf1  = (price * 100000) - ( material + labour + overhead)*100000

    # cashflow for year 2 
    cf2 = (price * 100000 * 1.05) - ( material*1.04 + labour*1.1 + overhead*1.07)*100000

    # cashflow for year 3
    cf3 = (price * 100000 * 1.1025) - ( material*1.0816 + labour*1.21 + overhead*1.1449)*100000
    
    npv = (cf1*0.856 + cf2*0.733 + cf3*0.628) - 2000000
    npv_.append(round(npv, 2))
```

```python
len(npv_)
```

```python
plt.figure(figsize=(15, 10))
plt.hist(npv_, bins=20)
plt.xlabel("Simulated Net Present Value")
plt.ylabel("Frequency")
plt.title("Monte Carlo simulations for Net Present Value of Sonic Cat")
```
